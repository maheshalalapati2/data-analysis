Basic Setup to load pandas and numpy libraries to perform all 4 fours joins: inner, left, right and outer

```python
import pandas as pd
import numpy as np
```

Creating 2 tables using dataframe with 4 key elements and 4 random values in each table. Here A, B and C ate the common keys.

```python
left_table =  pd.DataFrame({'key':['A','B','C','D'],'value':np.random.rand(4)})
right_table = pd.DataFrame({'key':['E','A','B','C'],'value':np.random.rand(4)})
```

Printing left table 

```python
left_table
```

printing right table

```python
right_table
```

INNER Join

```python
left_table.merge(right_table,on='key', how='inner')
```

LEFT join

```python
left_table.merge(right_table,on='key', how='left')
```

RIGHT join

```python
left_table.merge(right_table,on='key', how='right')
```

OUTER join

```python
left_table.merge(right_table,on='key', how='outer')
```
